## 2. The question: decision, action, cost of a wrong call
## 2. The Question

**Search question:** Which pages should a content team review first for refresh,
given limited review capacity?

**Unit of analysis:** one content page (content_hash_id), aggregated over a 90-day window.

**Output:** a ranked list of pages, each with a score, a suggested action
(refresh / expand / protect / monitor), and a reason code explaining why it was flagged.

**The decision this improves:** right now a reviewer either checks pages randomly or uses
a fixed hand-written rule. This model/ranking replaces "guessing" with an evidence-ranked
queue, so the team spends limited review hours on the highest-value pages first.

**The action someone takes:** a content editor opens the top N pages from the queue and
either rewrites/expands them, protects them from further decline, or leaves them for
monitoring — based on the reason code attached.

**Cost of a wrong recommendation:** if the model flags a page as high-priority and it
wasn't actually declining or losing demand, the team wastes editor hours on a page that
didn't need attention — and a genuinely declining page might get missed further down the
queue. This is a decision-support cost (wasted time / delayed action), not a safety-critical
one — but at scale, wasted review hours are real budget.

**Why data/ML helps at all:** with hundreds of thousands of content items and a handful of
reviewers, nobody can manually check every page. A hand-written rule (like the starter
baseline) already helps, but Notebook 02 showed a learned model beats it substantially at
picking the right pages first — meaning there's a pattern human-written rules miss.
